```python
from keras.layers import Conv2D,Convolution2D,MaxPooling2D,Activation
from keras.models import Sequential
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## 查看图片

```python
photo = cv2.imread('cat1.png')
plt.imshow(photo)
```

```python
#图片大小
photo.shape
```

## 一个卷积层的模型

```python
model = Sequential()
#3个3*3卷积核
model.add(Convolution2D(3,(3,3),input_shape=photo.shape))
```

```python
#为图象加入纵深维度，深度为1
photo_batch = np.expand_dims(photo,axis=0)
```

```python
#输入数据
conv_photo = model.predict(photo_batch)
```

```python
#输出图象
def show_conv_img(model,photo):
    photo_batch = np.expand_dims(photo,axis=0)
    conv_photo = model.predict(photo_batch)
    conv_photo = np.squeeze(conv_photo,axis=0)
    print(conv_photo.shape)
    plt.imshow(conv_photo)
```

```python
show_conv_img(model,photo)
```

## 3个10*10卷积核

```python
model = Sequential()
model.add(Convolution2D(3,(10,10),input_shape=photo.shape))
show_conv_img(model,photo)
```

## 1个卷积核

```python
def nice_photo_printer(model,photo):
    photo_batch = np.expand_dims(photo,axis=0)
    conv_photo2 = model.predict(photo_batch)
    conv_photo2 = np.squeeze(conv_photo2,axis=0)
    print(conv_photo2.shape)
    conv_photo2 = conv_photo2.reshape(conv_photo2.shape[:2])
    
    print(conv_photo2.shape)
    plt.imshow(conv_photo2)
```

```python
model = Sequential()
model.add(Convolution2D(1,(3,3),input_shape=photo.shape))
nice_photo_printer(model,photo)
```

```python
# 15*15
model = Sequential()
model.add(Convolution2D(1,(15,15),input_shape=photo.shape))
nice_photo_printer(model,photo)
```

## 激活函数

```python
model = Sequential()
model.add(Convolution2D(1,(3,3),input_shape=photo.shape))
#relu激活
model.add(Activation('relu'))
nice_photo_printer(model,photo)
```

```python
model = Sequential()
model.add(Convolution2D(3,(3,3),input_shape=photo.shape))
model.add(Activation('relu'))
show_conv_img(model,photo)
```

## 池化

```python
model = Sequential()
model.add(Convolution2D(1,(3,3),input_shape = photo.shape))
model.add(MaxPooling2D(pool_size=(5,5)))
nice_photo_printer(model,photo)
```

```python
model = Sequential()
model.add(Convolution2D(3,(3,3),input_shape = photo.shape))
model.add(MaxPooling2D(pool_size=(5,5)))
show_conv_img(model,photo)
```

## 激活+池化

```python
model = Sequential()
model.add(Convolution2D(1,(3,3),input_shape = photo.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(5,5)))
nice_photo_printer(model,photo)
```

```python
model = Sequential()
model.add(Convolution2D(3,(3,3),input_shape = photo.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(5,5)))
show_conv_img(model,photo)
```
